### Create a Word Cloud to see the most common words in the Review column

```python
#conda install wordcloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#FUNCTION TO DISPLAY WORDCLOUD OF DATA
def show_wordcloud(data, title = None):
    
    wordcloud = WordCloud(
        background_color = 'white',
        max_words = 200,
        max_font_size = 40, 
        scale = 3,
        random_state = 42
    ).generate(str(data))

    fig = plt.figure(1, figsize = (20, 20))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)

    plt.imshow(wordcloud)
    plt.show()
    
# print wordcloud
show_wordcloud(data["cleaned_review"])
```

#### This explains to us that they are fake reviews because from research, the word "free sample" is 90% of the time associated to a fake review or product.

#### Reading through the reviews, you will discover that some of the context in which the word "free sample" was used cannot be associated to a fake review. To solve this we check the ratings and only return the reviews with a rating greater than 3 as fake

```python
data['rating'] > 3
```

```python
df[df['rating'] > 3].count()
```

```python
def fake_review(text, rat):
    if 'free sample' in text and rat > 3:
        return 1
    else:
        return 0
```

```python
df = data[['cleaned_review', 'rating']].apply(lambda x:fake_review(x['cleaned_review'], x['rating']), axis=1)
df
```
